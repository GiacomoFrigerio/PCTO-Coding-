##Esercizio: moto del proiettile

L'altezza massima per il moto di un proiettile è data dalla formula:
$$ h_{max} = v_o^2 \, sin(\theta)^2 / 2 g$$

```python
import numpy as np

# Costante per l'accelerazione di gravità (in m/s^2)
g = 9.81

# Funzione per calcolare l'altezza massima raggiunta
def altezza_massima(v0, angolo):
    # Converti l'angolo in radianti
    angolo_rad = np.radians(angolo)

    # Calcola l'altezza massima utilizzando la formula
    h_max = (v0**2) * (np.sin(angolo_rad)**2) / (2 * g)

    return h_max

# Esempio di utilizzo della funzione
velocita_iniziale = 30  # in m/s
angolo_iniziale = 45    # in gradi

h = altezza_massima(velocita_iniziale, angolo_iniziale)
print("L'altezza massima raggiunta è: ",h, " metri")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Costante per l'accelerazione di gravità (in m/s^2)
g = 9.81

# Funzione per calcolare la traiettoria del proiettile
def traiettoria(v0, angolo):
    # Converti l'angolo in radianti
    angolo_rad = np.radians(angolo)

    # Tempo massimo di volo (quando il proiettile torna a terra)
    t_volo = 2 * v0 * np.sin(angolo_rad) / g

    # Crea un array di tempi da 0 al tempo massimo di volo
    t = np.linspace(0, t_volo, num=500)

    # Calcola le posizioni orizzontali (x) e verticali (y)
    x = v0 * np.cos(angolo_rad) * t
    y = v0 * np.sin(angolo_rad) * t - 0.5 * g * t**2

    return x, y

# Input di diverse velocità iniziali e angoli
velocita  = 30  # in m/s
angolo = 45    # in gradi

# Plot delle traiettorie
plt.figure()

x,y = traiettoria(velocita,angolo)
plt.plot(x, y, label=f"v0={velocita} m/s, angolo={angolo}°")

# Aggiunta di etichette e titolo al grafico
plt.title("Traiettoria del Proiettile per Diverse Velocità e Angoli")
plt.xlabel("Distanza orizzontale (m)")
plt.ylabel("Altezza (m)")
plt.legend()
#plt.grid(True)

# Mostra il grafico
plt.show()
```

##Spese mensili

```python
months = ["Jan","Feb","Mar","Apr", "May","June","July","Aug","Sept","Oct","Nov","Dec"]
affitto = [600, 600, 600, 600, 600, 600, 600, 600, 600, 600, 600, 600]
spesa = [300, 200, 350, 200, 180, 500, 300, 260, 270, 300, 280, 400]
utilities = [180, 150, 160, 200, 300, 190, 230, 180, 300, 210, 240, 360]

for i in range(len(spesa)):
 spesa[i] = spesa[i] + affitto[i]
 utilities[i] = utilities[i] + spesa[i] + affitto[i]

#plot in ordine, dal valore più alto al valore più basso
plt.bar(months, utilities, color="green", label = "utilities")
plt.bar(months, spesa,  color="blue", label = "spesa")
plt.bar(months,affitto , color="red", label = "affitto")

plt.legend()
plt.show()
```
